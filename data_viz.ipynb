```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 

df = pd.read_excel('excel\\all_sql_queries.xlsx', sheet_name = 3)

grouped_df = df.groupby('Product')['Total Revenue'].sum().reset_index()
grouped_df = grouped_df[grouped_df['Total Revenue'] >= 1000000]

plt.figure(figsize=(8, 6))
plt.pie(grouped_df['Total Revenue'], labels=grouped_df['Product'], autopct='%1.1f%%')
plt.title('Revenue Distribution by Product')
plt.axis('equal')
plt.show()
```
