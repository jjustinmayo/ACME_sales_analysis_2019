```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
sales = pd.read_csv('csv\\acme_combined_sales_2019.csv')
sales.head()
```

## Product Revenue

```python
sales['Total Revenue'] = sales['Quantity Ordered'] * sales['Price Each']
sales.head()
```

Use a bar graph to graph the top 10 products by revenue 

```python
sales_product = sales.groupby('Product')['Total Revenue'].sum().sort_values(ascending=False)

plt.figure(figsize=(14,6))
sales_product.head(10).plot(title="Product Sales", label="", kind='bar')
plt.show()
```

Graph the top 10 products by revenue using a pie chart

```python
plt.figure(figsize=(8,8))
sales_product.head(10).plot(title='Product Sales', label="", kind='pie')
plt.show()
```
